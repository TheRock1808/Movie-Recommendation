```python
data = pd.merge(new_meta, credit, on='id')
```

```python
print(data['genres'].apply(type).unique())
print(data['cast'].apply(type).unique())
print(data['crew'].apply(type).unique())
```

```python
# to convert string into list 
import ast
data['cast'] = data['cast'].map(lambda x : ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x : ast.literal_eval(x))
```

```python
# to separate genres indiviuals
def make_genrelist(x):
    if not isinstance(x, list):  # Ensure x is a list before processing
        return np.NaN
    gen = []
    st = " "
    for i in x:
        if i['name'] == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i['name'])
    return np.NaN if not gen else st.join(gen)
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_genrelist(x))
```

```python
# retrieve actors and director name from the list 
def get_actor1(x):
    casts = []
    st = " "
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    return (casts[0])

def get_actor2(x):
    casts = []
    st = " "
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts) <= 1:
        return np.NaN
    return (casts[1])

def get_actor3(x):
    casts = []
    st = " "
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts) <= 2:
        return np.NaN
    return (casts[2])

def get_director(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    return (st.join(dt))
```

```python
data['actor_1_name'] = data['cast'].map(lambda x : get_actor1(x))
data['actor_2_name'] = data['cast'].map(lambda x : get_actor2(x))
data['actor_3_name'] = data['cast'].map(lambda x : get_actor3(x))
data['director_name'] = data['crew'].map(lambda x : get_director(x))
```

```python
movie = data.loc[:, ['director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'genres_list', 'title']]
```

```python
movie.isna().sum()
```

```python
# drop NaN values 
movie = movie.dropna(how='any')
```

```python
movie = movie.rename(columns={'title':'movie_title'})
movie = movie.rename(columns={'genres_list':'genres'})
```

```python
movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' ' + movie['actor_3_name'] + ' ' + movie['director_name'] + ' ' + movie['genres']
```

```python
old = pd.read_csv('data.csv')
```

```python
new = pd.concat([old, movie])
```
