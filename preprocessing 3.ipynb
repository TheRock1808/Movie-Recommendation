```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Dataset/All_Movies1.csv', lineterminator='\n')
```

```python
data.head()
```

```python
data = data.loc[:, ['Title', 'Director', 'Cast', 'genres']]
```

```python
def split_actors(row):
    actors = row.split(',')
    actors = [actor.strip() for actor in actors]
    
    actors += [np.nan] * (3 - len(actors))
    
    return pd.Series(actors[:3], index=['actor_1_name', 'actor_2_name', 'actor_3_name'])

# Apply the function to the DataFrame
data[['actor_1_name', 'actor_2_name', 'actor_3_name']] = data['Cast'].apply(split_actors)
```

```python
data.drop(columns=['Cast'], inplace=True)
```

```python
data['genres'] = data['genres'].str.replace(',', ' ')
```

```python
data = data.rename(columns={'Title':'movie_title'})
data = data.rename(columns={'Director':'director_name'})
```

```python
data['comb'] = data.apply(
    lambda row: (
        str(row['actor_1_name']) + ' ' + str(row['director_name']) + ' ' + str(row['genres'])
        if pd.isna(row['actor_2_name']) 
        else str(row['actor_1_name']) + ' ' + str(row['actor_2_name']) + ' ' + str(row['actor_3_name']) + ' ' + str(row['director_name']) + ' ' + str(row['genres'])
    ), axis=1
)
```

```python
data
```

```python
data['actor_1_name'] = data['actor_1_name'].replace(np.nan, 'unknown')
data['actor_2_name'] = data['actor_2_name'].replace(np.nan, 'unknown')
data['actor_3_name'] = data['actor_3_name'].replace(np.nan, 'unknown')
data['director_name'] = data['director_name'].replace(np.nan, 'unknown')
```

```python
data
```

```python
data['movie_title'] = data['movie_title'].str.lower()
```

```python
old = pd.read_csv('new_data.csv')

new_data_1 = pd.concat([old, data])
```

```python
new_data_1
```

```python
new_data_1.drop_duplicates(subset = "movie_title", keep = 'last', inplace = True)
```
